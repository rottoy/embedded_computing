```python
#include <Python.h>
#include <wiringPi.h>
#define DEVICE_FILENAME  "/dev/Ex_06"    
#include "simple.h"

static PyObject *blink(PyObject *self, PyObject *args)
{
    int ledPin = 0, pinNum = 0;
    int i = 0; // for
	ledHW ledhw;
    
    if(!PyArg_ParseTuple(args, "i", &pinNum))
        return NULL;
	
    
    wiringPiSetupGpio();
    pinMode(ledPin, OUTPUT);
    
    for(i=0; i<5; i++){
        digitalWrite(ledPin, HIGH);
        delay(500);
        digitalWrite(ledPin, LOW);
        delay(500);
    }
    Py_RETURN_NONE;
}

static PyObject *led(PyObject *self, PyObject *args) {
	int pin = 0;
	int order = 0;
	int fd = 0;
	int *inputPin;
	char *inputOrder;

	if (!PyArg_ParseTuble(args, "i s", &inputPin,&inputOrder))
		return NULL;
	memset(&ledhw, 0, sizeof(ledhw));
	int size = sizeof(ledhw);
	ledPin = pinNum;
	pin = inputPin;
	if (strcmp(inputOrder, "1"))
		order = 1;
	else
		order = 0;
	printf("%d, %d 입니다.", pin, order);
	ledhw.pin = pin;
	ledhw.funcNum = order;
	fd = open(DEVICE_FILENAME, O_RDWR | O_NDELAY);
	if (fd >= 0) {
		printf("send ledctl to device dirver\n");
		ioctl(fd, MY_IOC_SET, &ledhw);
		ioctl(fd, MY_IOC_GPIO_SETFUNC);
	}
	else
		printf("코드다시짜임마");
	close(fd);

	return NULL;
	

}
static PyMethodDef keywdarg_methods[] = {
    {"blink", (PyCFunction)blink, METH_VARARGS, NULL},
    {NULL, NULL, 0, NULL}   /* sentinel */
};

static struct PyModuleDef keywdargmodule = {
    PyModuleDef_HEAD_INIT,
    "blink",
    NULL,
    -1,
    keywdarg_methods
};

PyMODINIT_FUNC PyInit_ex3(void)
{
    return PyModule_Create(&keywdargmodule);
}

```
